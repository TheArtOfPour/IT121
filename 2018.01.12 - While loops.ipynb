```python
fruits = ["banana", "papaya", "starfruit", "mango", "iphone X"]

# for loops use iterators to count or step through collections
for fruit in fruits:
    print("I heard that guy really likes %s\n" % fruit)
    
print("--------------------------------------------------")

# while loops operate on conditions (boolean checks)
index = 0
while index < len(fruits):
    print("While I heard that they don't like %s\n" % fruits[index])
    index += 1
```

```python
import random
shape = input("What's your favorite shape?")
allowed_shapes = ["circle", "square", "tesseract", "iphone X"]

count = 0
while not shape in allowed_shapes:
    if count < 2:
        shape = input("Pick another shape...(circle|square|tesseract)")
    elif count < 4:
        shape = input("Really? Just type 'c' 'i' 'r' 'c' 'l' 'e'")
    elif count < 5:
        shape = input('''Okay, you know what, last chance. 
                      Mess this up and you get what I pick.''')
    else:
        random_index = random.randint(0, len(allowed_shapes)-1)
        shape = allowed_shapes[random_index]
    count += 1

print("Oh %s, that's a good shape..." % shape)
```

```python
a = 10
b = 20
c = 5
count = 0
while a > 0:
    while a+b-c != 0:
        while c < 10:
            count += 1
            c += 2
        if a+b-c > 0:
            b -= 1
        else:
            b += 1
    a -= 1
print("Final countdown %d\n" % count)
```
