```python
# @todo: get board_size from user input
import random
board_size = int(input("What board size would you like? "))
board = []
tokens = ['X', 'O']
player = random.randint(0, 1)
print("Player %d starts %s" % ((player+1), tokens[player]))

def buildBoard():
    for _ in range(board_size):
        new_column = []
        for _ in range(board_size):
            new_column.append(' ')
        board.append(new_column)

def drawBoard():
    horizontal_line = "\n" + (board_size-1)*"-+" + "-"
    print("\n")
    for i, col in enumerate(board):
        for j, cell in enumerate(col):
            print(cell, end='')
            if j < len(col) - 1:
                print('|', end='')
        if i < len(board) - 1:
            print(horizontal_line)
    print("\n")

def move(row, col, token):
    if board[row][col] == ' ':
        board[row][col] = token
        return True
    return False

# @todo : check for winner
def didWin():
    # ROW CHECK
    for i, col in enumerate(board):
        did_win = True
        token_check = ""
        for j, cell in enumerate(col):
            if board[i][j] == " ":
                did_win = False
                break
            if token_check == "":
                token_check = board[i][j]
            if token_check != board[i][j]:
                did_win = False
                break
            print("(%d, %d)" % (i, j))
        if did_win:
            return True
        
    # COLUMN CHECK
    for i, col in enumerate(board):
        did_win = True
        token_check = ""
        for j, cell in enumerate(col):
            if board[j][i] == " ":
                did_win = False
                break
            if token_check == "":
                token_check = board[j][i]
            if token_check != board[j][i]:
                did_win = False
                break
            print("(%d, %d)" % (i, j))
        if did_win:
            return True
        
    return False

buildBoard()
game_over = False
while not game_over:
    # Get player input
    row = int(input("Which row? (1-%d)" % board_size))-1              
    if row == -1:
        break
    col = int(input("Which col? (1-%d)" % board_size))-1
    
    # Submit move
    token = tokens[player]
    success = move(row, col, token)
    if not success:
        print("Invalid move\n")
        continue
    
    # Check for win
    if didWin():
        drawBoard()
        print("Player %d Wins!!!!!!!!" % (player+1))
        break
    
    # Toggling player
    if player == 0:
        player = 1
    else:
        player = 0
    drawBoard()
```
